# 1. 2019 data

```python
import geopandas as gpd
import pandas as pd

from shapely.geometry import Point

import jenkspy as jk
import pandas as pd

import warnings
warnings.simplefilter('ignore') 
```

```python
DATA_PATH = "./data/raw_data/Accident data/"
MAP_PATH = "./data/Map/LSOA/"
FILE_NAME = "Accidents_2019.csv"

# obtain accidents data in 2019
raw_data = pd.read_csv(DATA_PATH+FILE_NAME)
accidents_19 = gpd.GeoDataFrame(raw_data, geometry = [Point(xy) for xy in zip (raw_data['Location_Easting_OSGR'], raw_data['Location_Northing_OSGR'])])
```

```python
# open London lsoa polygon shapefile
lsoas_gdf = gpd.read_file('./data/map/LSOA_accident.shp')
lsoas_gdf
#lsoas_gdf.to_file(driver='ESRI Shapefile', filename='./data/Map/LSOA/lsoa_accidents.shp')
```

```python
# select all road traffic accidents in London in 2019
PointInPolys = gpd.sjoin(lsoas_gdf, accidents_19, op = 'intersects', how='left')
accident_gdf = gpd.GeoDataFrame(PointInPolys, geometry = [Point(xy) for xy in zip (PointInPolys['Location_Easting_OSGR'], PointInPolys['Location_Northing_OSGR'])])
accident_gdf.dropna(axis = 0, inplace = True)
accident_gdf.to_file(driver='ESRI Shapefile', filename='./data/map/london_accidents_2019.shp')
```

## Data processing for the geodetector

```python
lsoa_gdf_2019 = gpd.read_file("./data/geodetector/2019/accidents_swi_2019.shp")
lsoa_gdf = gpd.read_file("./data/geodetector/nbregression_data_lsoa.shp")
```

```python
data_poi = lsoa_gdf[['code', 'lsoa', 'borough', 
       'ACCOMM', 'EDU', 'FOOD','PARK', 'SHOP', 'TRANSPORT', 'WORK', 
       'ENTERTAIN', 'ATTRACT']]
```

```python
data_poi.FOOD.sum()
```

```python
data_swi = lsoa_gdf_2019[['code', 'SWI', 'count']]
```

```python
# data preparation for geodetector
data_gdtr = pd.merge(data_poi, data_swi, how='left',on='code')
```

```python
# JNB classification
class_nb = 5
poi = ['EDU', 'FOOD', 'SHOP', 'TRANSPORT', 'WORK', 'ENTERTAIN', 'ATTRACT']
for i in poi:
    breaks = jk.jenks_breaks(data_gdtr[i],nb_class=class_nb)
    k = 0
    data_gdtr[i][data_gdtr[i]>=breaks[0]]
    data_gdtr[i+'_jnb']=data_gdtr[i]
    for n in breaks:
        k += 1
       
        if k < class_nb+1:
            print(k)
            data_gdtr[i+'_jnb'][data_gdtr[i]>=n]=k
        else:
            break
    print(breaks)
```

```python
class_nb = 3
poi = ['ACCOMM','PARK']
for i in poi:
    breaks = jk.jenks_breaks(data_gdtr[i],nb_class=class_nb)
    k = 0
    data_gdtr[i][data_gdtr[i]>=breaks[0]]
    data_gdtr[i+'_jnb']=data_gdtr[i]
    for n in breaks:
        k += 1
       
        if k < class_nb+1:
            print(k)
            data_gdtr[i+'_jnb'][data_gdtr[i]>=n]=k
        else:
            break
    print(breaks)
```

```python
# all geodetector data
data_gdtr.to_csv('./data/geodetector/2019/all/geodetector_data_2019.csv', index=False)
```

```python
# weekday geodetector data
data_wd = pd.read_csv('./data/geodetector/2019/weekday/patterns_weekday.csv')
data_wd = data_wd[['code', 'Topic 1', 'Topic 2',
       'Topic 3', 'Topic 4', 'Topic 5', 'Topic 6', 'Topic 7', 'Topic 8',
       'Topic 9', 'Topic 10', 'Topic 11', 'Topic 12', 'Topic 13']]

# data preparation for geodetector
data_gdtr_wd = pd.merge(data_gdtr, data_wd, how='left',on='code')

data_gdtr_wd.to_csv('./data/geodetector/2019/weekday/geodetector_data_weekday19.csv', index=False)
```

```python
# weekend geodetector data
data_we = pd.read_csv('./data/geodetector/2019/weekend/patterns_weekend.csv')
data_we = data_we[['code', 'Topic 1', 'Topic 2',
       'Topic 3', 'Topic 4', 'Topic 5', 'Topic 6', 'Topic 7', 'Topic 8']]

# data preparation for geodetector
data_gdtr_we = pd.merge(data_gdtr, data_we, how='left',on='code')

data_gdtr_we.to_csv('./data/geodetector/2019/weekend/geodetector_data_weekend19.csv', index=False)
```

# 2. 2020 data

```python
import geopandas as gpd
import pandas as pd

from shapely.geometry import Point

import jenkspy as jk
import pandas as pd

import warnings
warnings.simplefilter('ignore') 
```

```python
DATA_PATH = "./data/raw_data/Accident data/"
MAP_PATH = "./data/Map/LSOA/"
FILE_NAME = "Accidents_2020.csv"

# obtain accidents data in 20
raw_data = pd.read_csv(DATA_PATH+FILE_NAME)
accidents_20 = gpd.GeoDataFrame(raw_data, geometry = [Point(xy) for xy in zip (raw_data['location_easting_osgr'], raw_data['location_northing_osgr'])])
```

```python
# open London lsoa polygon shapefile
lsoas_gdf = gpd.read_file('./data/map/LSOA_accident.shp')
lsoas_gdf
#lsoas_gdf.to_file(driver='ESRI Shapefile', filename='./data/Map/LSOA/lsoa_accidents.shp')
```

```python
# select all road traffic accidents in London in 2010-2019
PointInPolys = gpd.sjoin(lsoas_gdf, accidents_20, op = 'intersects', how='left')
accident_gdf = gpd.GeoDataFrame(PointInPolys, geometry = [Point(xy) for xy in zip (PointInPolys['location_easting_osgr'], PointInPolys['location_northing_osgr'])])
accident_gdf.dropna(axis = 0, inplace = True)
accident_gdf.to_file(driver='ESRI Shapefile', filename='./data/map/london_accidents_2020.shp')
```

```python
accident_gdf
```

```python
# drop the duplicated accidents based on accident ID.
def drop_duplicated_accidents(path):
    raw_data = gpd.read_file(path)
    data = raw_data.explode()
    data = data.drop(['level_0','level_1'],axis=1)
    accidents = data.drop_duplicates(['accident_i'])
    accidents.to_file(driver='ESRI Shapefile', filename=path)
    return 'The length of raw data is {}, and there are {} duplicated accidents.'.format(len(raw_data.accident_i), len(raw_data.accident_i)-len(accidents.accident_i))

accident_path = 'F:\\Individual programme\\data\\London\\accidents\\accident_2020\\snap_accidents_2020.shp'
drop_duplicated_accidents(accident_path)
```
